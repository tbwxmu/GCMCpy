```python
s = open('step1_pdbreader.pdb', 'r').read()
```

```python
p = s.strip().split('\n')
```

```python
cryst = None
for line in p:
    if line.lower().startswith('cryst1 '):
        cryst = line.split()[1:4]
        break
cryst
```

```python
atoms = []
for line in p:
    if (line.lower().startswith('atom ') or line.lower().startswith('hetatm')) and len(line) > 54:
        x = float(line[30:38])
        y = float(line[38:46])
        z = float(line[46:54])
        serial = line[6:11].strip()
        name = line[12:16].strip()
        residue = line[17:20].strip()
        sequence = line[22:26].strip()
        atoms.append((serial, name, residue, sequence, x, y, z, line))
len(atoms)
```
